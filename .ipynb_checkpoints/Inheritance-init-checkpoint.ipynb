```python
class Person(object):
    def __init__(self,name):
        self.name = name
    def getName(self):
        return self.name
    def isEmployee(self):
        return False
```

```python
class Employee(Person):
    def isEmployee(self):
        return True
```

```python
emp = Person("Anshu")
print(emp.getName(), emp.isEmployee())
```

```python
emp = Employee("Amit")
print(emp.getName(), emp.isEmployee())
```
